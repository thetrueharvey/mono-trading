```python
from datetime import datetime
```

```python
from polars import DataFrame, col, Float32
```

```python
pair_df = (
    DataFrame(
        {
            'timestamp': [datetime(2020, 1, 1, 1, x) for x in range(60)],
            'open': [float(x) for x in range(60)],
            'high': [float(x) for x in range(60)],
            'low': [float(x) for x in range(60)],
            'close': [float(x) for x in range(60)],
        },
    )
    .with_columns(col(['open', 'high', 'low', 'close']).cast(Float32))
    .write_parquet('trading/data/test/resources/simple_pair_df.parquet')
)
```
