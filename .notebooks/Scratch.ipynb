# General Scratch Pad

```python
import polars as pl
```

```python
from trading.utils.Pairs import Asset, Pair
```

```python
(df := pl.read_parquet('.data/binance/exchange.parquet')).head()
```

```python
assets = {}
for asset in [*df['baseAsset'].unique()] + [*df['quoteAsset'].unique()]:
    assets[asset] = Asset(name=asset)
```

```python
(
    df := (
        df
            .with_columns(
                [
                pl.col('baseAsset').apply(lambda x: assets[x], return_dtype=pl.Object),
                pl.col('quoteAsset').apply(lambda x: assets[x], return_dtype=pl.Object),
                ]
            )
    )
).head()
```

```python
(
    df := (
        df.with_column(
            pl.Series(
                values=[Pair(asset=asset, base=base) for asset, base in zip(df['baseAsset'], df['quoteAsset'])],
                name='Asset'
            )
        )
    )
).head()
```
